```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Data for Youngs (lengths in nm)
d = 1000.0     # slit separation: 1 um
lamb = 650.0   # wavelength of laser

# Data for plotting
N = 4 # number of slits
theta = np.linspace(-np.pi/2.0, np.pi/2.0, 3141) # graph as a function of theta
delta = (2.0 * np.pi / lamb) * d * np.sin(theta)

# scaling factors as to ensure same maximum intensity
bigI = (np.sin(N*delta / 2.0) / np.sin(delta / 2.0))**2 # intensity
bigI2 = 1/16 * (np.sin((N**2)*delta / 2.0) / np.sin(delta / 2.0))**2
bigI3 = 1/4096 * (np.sin((N**4)*delta / 2.0) / np.sin(delta / 2.0))**2 

dovlamsintheta = (d / lamb) * np.sin(theta)
fig, ax = plt.subplots()
ax.plot(dovlamsintheta, bigI, color='r')
ax.plot(dovlamsintheta, bigI2, color='b')
ax.plot(dovlamsintheta, bigI3, color='black')
ax.legend(['N = 4', 'N = 16', 'N = 256'],
          loc='center right',
          title = "Number of slits",
          bbox_to_anchor=(1.3, 0.5))

ax.set(xlabel='position (d/lambda*sin(theta)', ylabel='Intensity (I)', title='Youngs Double Slit Intensity')
ax.grid()

fig.savefig("youngs_manyslits.png")
plt.show()
```

```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# lengths in nm
a = 1500.0     # slit width: 1 um
lamb1 = 350.0   # wavelength of blue laser
lamb2 = 700.0  # wavelength of red laser

# Data for plotting
theta = np.linspace(- np.pi / 2, np.pi / 2, 1000) # graph as a function of theta
beta1 = (np.pi / lamb1) * a * np.sin(theta)
beta2 = (np.pi / lamb2) * a * np.sin(theta)

bigI1 = (np.sin(beta1) / beta1)**2 # intensity for single slit diffraction
bigI2 = (np.sin(beta2) / beta2)**2

fig, ax = plt.subplots()
ax.plot(theta, bigI1, color='r')
ax.plot(theta, bigI2, color='b')
ax.legend(['Red light', 'Blue light'],
          loc='center right',
          bbox_to_anchor=(1.3, 0.5))

ax.set(xlabel='Theta', ylabel='Intensity (I)', title='Single-Slit Diffraction')
ax.grid()

fig.savefig("singleslit.png")
plt.show()
```

Graph 3: plotting the single-slit diffraction patterns for two wavelengths of red (700 nm) and blue (350 nm) light. Slit width: 1.5 um.

```
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# lengths in nm
a = 1400.0     # slit width: 1.4 um
lamb = 700.0  # wavelength of red laser
d = a*3        # slit separation


# Single Slit Diffraction
theta = np.linspace(- np.pi / 2, np.pi / 2, 1000) # graph as a function of theta
beta1 = (np.pi / lamb) * a * np.sin(theta)
bigID = (np.sin(beta1) / beta1)**2 # intensity for single slit diffraction

# Young's Double Slit
delta = (2.0 * np.pi / lamb) * d * np.sin(theta)
bigII = 1/4 * (np.sin(2*delta / 2.0) / np.sin(delta / 2.0))**2 # intensity

# Together
together = bigII*bigID

fig, ax = plt.subplots()
ax.plot(theta, bigID, color='r')
ax.plot(theta, bigII, color='mediumpurple')
ax.plot(theta, together, color='black')
ax.legend(['Diffraction', 'Interference', 'Together'],
          loc='center right',
          bbox_to_anchor=(1.5, 0.5))

ax.set(xlabel='Theta', ylabel='Normalized Intensity (I)', title='Single-Slit Diffraction')
ax.grid()

fig.savefig("singleslit.png")
plt.show()
```
