```
# Alexa Jakob
# Modern Physics, Ph214-S: Prof Debroy
# HW 2, Due February 13, 2020

import matplotlib
import matplotlib.pyplot as plt
import numpy as np

#Parameters, assume z = 0
eps = 0
lamb = 670 # [nm]
c = 3e8 #[m/s, speed of light]
w = 2*np.pi*c/lamb*10-7
E0x = 2 # arbitrary scaling to show differences
E0y = 5 # arbitrary scaling to show differences

t = np.linspace(0, 4*np.pi/w)
Ex = E0x*np.cos(w*t+eps)
Ey = E0y*np.cos(w*t)
fig, ax = plt.subplots()
ax.plot(t, Ex, color='r')
ax.plot(t, Ey, color='b', linestyle = '--')

ax.legend(['Ex', 'Ey'],
          title = "Component",
          bbox_to_anchor=(1.3, 0.5))
ax.set(xlabel='Time (t)', ylabel='Electric field component, scaled', title='X and Y Components of Electric Wave (no phase difference)')
ax.grid()
```

Note that the two components are in phase and are identical except for amplitude, scaling factors E0x and E0y.

```
t = np.linspace(0, 4*np.pi/w)
Ex = E0x*np.cos(w*t+eps)
Ey = E0y*np.cos(w*t)
fig, ax = plt.subplots()
ax.plot(Ex, Ey, color='r')

ax.set(xlabel='Ex(t)', ylabel='Ey(t)', title='Variation of Electric Field Vector Tip Over Time (no phase difference)')
ax.grid()
```

Since epsilon = 0, we have a line plotted: it is plane polarization. The tilt is pi/4.

```
eps = np.pi
t = np.linspace(0, 4*np.pi/w)
Ex = E0x*np.cos(w*t+eps)
Ey = E0y*np.cos(w*t)
fig, ax = plt.subplots()
ax.plot(t, Ex, color='r')
ax.plot(t, Ey, color='b', linestyle = '--')

ax.legend(['Ex', 'Ey'],
          title = "Component",
          bbox_to_anchor=(1.3, 0.5))
ax.set(xlabel='Time (t)', ylabel='Electric field component, scaled', title='X and Y Components of Electric Wave (phase: pi)')
ax.grid()
```

```
#Parameters, assume z = 0
eps = np.pi
lamb = 670 # [nm]
c = 3e8 #[m/s, speed of light]
w = 2*np.pi*c/lamb*10-7
E0x = 2 # arbitrary scaling to show differences
E0y = 5 # arbitrary scaling to show differences

t = np.linspace(0, 4*np.pi/w)
Ex = E0x*np.cos(w*t+eps)
Ey = E0y*np.cos(w*t)
fig, ax = plt.subplots()
ax.plot(Ex, Ey, color='r')

ax.set(xlabel='Ex(t)', ylabel='Ey(t)', title='Variation of Electric Field Vector Tip Over Time (phase: pi)')
ax.grid()
```

Since epsilon = pi, we have a plane polarization here as well. The tilt is -pi/4.

```
eps = np.pi/2
t = np.linspace(0, 4*np.pi/w)
Ex = E0x*np.cos(w*t+eps)
Ey = E0y*np.cos(w*t)
fig, ax = plt.subplots()
ax.plot(t, Ex, color='r')
ax.plot(t, Ey, color='b', linestyle = '--')

ax.legend(['Ex', 'Ey'],
          title = "Component",
          bbox_to_anchor=(1.3, 0.5))
ax.set(xlabel='Time (t)', ylabel='Electric field component, scaled', title='X and Y Components of Electric Wave (phase: pi/2)')
ax.grid()
```

```
t = np.linspace(0, 4*np.pi/w)
Ex = E0x*np.cos(w*t+eps)
Ey = E0y*np.cos(w*t)
fig, ax = plt.subplots()
ax.plot(Ex, Ey, color='r')
ax.axis('equal')

ax.set(xlabel='Ex(t)', ylabel='Ey(t)', title='Variation of Electric Field Vector Tip Over Time (phase: pi/2)')
ax.grid()
```

The phase difference here is pi/2, which means the parametric graph yields an ellipse. Our scaling factors determine the length of the axes of the ellipse. In this case, we obtain elliptical polarization.

```
# Set E0x = E0y = 1
t = np.linspace(0, 4*np.pi/w)
Ex = np.cos(w*t+eps)
Ey = np.cos(w*t)
fig, ax = plt.subplots()
ax.plot(Ex, Ey, color='r')
ax.axis('equal')

ax.set(xlabel='Ex(t)', ylabel='Ey(t)', title='Variation of Electric Field Vector Tip (phase: pi/2, E0x = E0y)')
ax.grid()
```

This is a circle, because the phase is pi/2 and because the amplitude scaling constants E0x and E0y are equal. Therefore, we obtain circular polarization.
